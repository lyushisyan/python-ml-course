```python
import matplotlib.pyplot as plt
import numpy as np
```

## 1. Построение линейного графика (绘制折线图)

```python
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

### 1.1 Добавить новую линию и легенды (增加新的图线和图例)

```python
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')

plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```

### 1.2 Изменить размер изображения (修改图像尺寸)

```python
plt.figure(figsize=(6, 4), dpi=100, facecolor='white')

plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')

plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```

### 1.3 Изменить цвета, толщину линий и размер шрифта (修改颜色，线宽，字体大小)

```python
plt.figure(figsize=(6, 4), dpi=100, facecolor='white')

plt.plot(x, y1, label='sin(x)', color='red', linestyle='-', linewidth=2)
plt.plot(x, y2, label='cos(x)', color='black', linestyle='--', linewidth=2)

plt.xlabel('x', fontsize=18)
plt.ylabel('y', fontsize=18)
plt.tick_params(labelsize=16)
plt.legend(fontsize=16)
plt.grid()

plt.savefig("plot.png", bbox_inches='tight', dpi=300)
plt.show()
```

## 2. Построение диаграммы рассеяния (绘制散点图)

```python
x = np.linspace(0, 10, 50)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(6, 4), dpi=100)

plt.scatter(x, y1, color='blue', label='sin(x)')
plt.plot(x, y2, color='green', marker='o',
         linestyle='-', linewidth=2,
         label='cos(x)')

plt.xlabel('x', fontsize=18)
plt.ylabel('y', fontsize=18)
plt.tick_params(labelsize=16)
plt.legend(fontsize=14)
plt.grid()

plt.savefig("scatter.png", bbox_inches='tight', dpi=300)
plt.show()
```

## 3. Построение гистограммы (绘制直方图)

```python
data = np.random.normal(loc=0, scale=1, size=1000)

plt.figure(figsize=(6, 4), dpi=100)

plt.hist(data, bins=30, 
    color='skyblue', 
    edgecolor='black',
    alpha=0.8, label='N(0,1)')

plt.xlabel('Значения', fontsize=16)
plt.ylabel('Частота', fontsize=16)
plt.tick_params(labelsize=14)
plt.legend(fontsize=14)

plt.tight_layout()
plt.savefig("histogram.png", bbox_inches='tight', dpi=300)
plt.show()
```

## 4. Построение двумерной карты (构建二维云图)

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

Z = np.sin(np.sqrt(X**2 + Y**2))

plt.figure(figsize=(6, 5))
contour = plt.contourf(X, Y, Z, levels=50, cmap='viridis')

plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)
plt.title('Contourf Plot', fontsize=18)
plt.tick_params(labelsize=14)
cbar = plt.colorbar(contour)
cbar.set_label('z', fontsize=16)
cbar.ax.tick_params(labelsize=14)

plt.grid(True)
plt.tight_layout()
plt.show()
```

## 5. Построение трёхмерной поверхности

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

fig = plt.figure(figsize=(7, 5))
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('x', fontsize=14)
ax.set_ylabel('y', fontsize=14)
ax.set_zlabel('z', fontsize=14)

plt.tight_layout()
plt.show()
```
