## 2. Структура данных (数据结构)

### 2.1 список (list) (列表)
- Упорядоченная коллекция  (有序集合)
- Можно добавлять, удалять и изменять элементы  (可以添加、删除和修改元素)
- Можно хранить разные типы данных (числа, строки и т.д.)  (可以存储不同的数据类型（数值、字符串等）)
- Можно использовать повторяющиеся значения  （可以使用重复值）
- Можно обращаться к элементам по индексу и делать срезы  （可以通过索引访问元素并执行切片操作）
- Записывается в в квадратных скобках: [ ]  （用方括号 [ ] 表示）

```python
lst = [1, 2, 3, 4]
lst + 1
```

```python
lst.append(5)  # добавить в конец (添加到末尾)
lst
```

```python
lst.insert(1, 10)  # вставить на позицию (插入至某处)
lst
```

```python
lst.remove(3)  # удалить первое вхождение (删除第一个实例)
lst
```

```python
lst.pop(3)  # удалить и вернуть элемент по индексу (根据索引删除并返回该元素)
```

```python
lst # Список после удаления элементов (删除元素后的列表)
```

```python
lst[1:3]   # срез от i до j (не включая j) (从 i 到 j（不包括 j）的切片)
```

- Упорядоченная коллекция, как список, но нельзя изменять (有序集合，类似于列表，但不可修改)

- Можно использовать повторяющиеся значения (可以使用重复值)
- Можно обращаться к элементам по индексу и делать срезы (可以通过索引访问元素并执行切片操作)

```python
tpl = (1, 2, 3, 4)
```

```python
tpl[1:3]  # срез (подкортеж) (切片（子元组）)
```

```python
len(tpl)  # длина кортежа (元组长度)
```

- Обозначается фигурными скобками: {} (用花括号 {} 表示)

## 3. Функция (函数)
### 3.1 Пользовательская функция (自定义函数)
Используется ключевое слово *def* для определения своей функции (关键字 *def* 用于定义函数)

```python
# Пользовательская функция: среднее значение списка
# 自定义函数：计算列表中所有元素的平均值
def average(lst):
    return sum(lst) / len(lst)

a = [2, 3, 1, -4, 12, 6]
print(average(a))
```

```python
# Пользовательская функция: считаем количество положительных элементов
# 自定义函数：统计列表中 > 0 的元素个数
def count_positive(lst):
    count = 0
    for x in lst:
        if x > 0:
            count += 1
    return count

a = [2, 3, 1, -4, 12, 6, 0, -3]
print(count_positive(a))
```

### 3.2 Анонимная функция（匿名函数）
Краткая запись простой функции (用于书写简单的小函数)

```python
# Квадратная функция: f(x) = x**2
# 定义一个平方函数
square = lambda x: x ** 2
print(square(5))
```

```python
# Функция map применяет lambda к каждому элементу списка
# map 会把 lambda 函数应用到列表的每一个元素上
nums = [1, 2, 3, 4]
squared = list(map(lambda x: x ** 2, nums))
print(squared)
```

## 4. Цикл (循环)
### 4.1 Цикл for（for 循环）
Используется для перебора объектов, из которых можно получать элементы по одному  
用于遍历那些能一个一个取出元素的对象（列表、元组、字符串、字典、集合）

#### 4.1.1 Перебор списка (遍历列表) :

```python
nums = [1, 2, 3, 4]
for num in nums:
    print(num)
```

#### 4.1.2 Перебор строки (遍历字符串):

```python
text = "abcd"
for char in text:
    print(char)
```

#### 4.1.3 Перебор словаря (遍历字典):

```python
dic = {'a': 1, 'b': 2, 'c': 3}
for key, value in dic.items():
    print(f"{key}: {value}")
```

#### 4.1.4 Использование range (range的使用):

```python
for i in range(4):
    print(i)
```

### 4.2 Цикл while（while 循环）
Выполняется, пока условие остаётся True. Если условие становится False, цикл завершается
(当条件为 True 时循环持续执行；一旦为 False，循环结束)

#### 4.2.1 Счётный цикл (计数循环)

```python
count = 0
while count < 5:
    print(count)
    count += 1
```

#### 4.2.2 Условный цикл (条件循环)

```python
n = 10
while n > 0:
    print(n)
    n -= 2
```

### 4.3 Операторы управления циклами (循环控制语句)
#### 4.3.1 break — досрочно прерывает цикл (提前结束循环)
Используется, когда нужно немедленно выйти из цикла (用于立即跳出循环)

```python
for i in range(10):
    if i == 5:
        break
    print(i)
```

#### 4.3.2 continue — перейти к следующей итерации (跳过当前循环，继续下一轮)
Используется, когда нужно пропустить часть кода, но цикл продолжить (用于跳过本轮循环剩余代码，进入下一轮)

```python
for i in range(5):
    if i == 2:
        continue
    print(i)
```

#### Задание 2. Найдите наибольшее произведение двух разных элементов списка (找到列表中两个不同元素的最大乘积)

Пример списка: `nums = [1, 5, 3, 9, 2]`

Ожидаемый вывод: `45`

Решение：

```python
nums = [1, 5, 3, 9, 2]

max_product = 0
for i in range(len(nums)):
    for j in range(i + 1, len(nums)):
        product = nums[i] * nums[j]
        if product > max_product:
            max_product = product

print("Максимальное произведение:", max_product)
```

#### Задание 3. Подсчитать количество чётных чисел в списке (统计列表中有多少个偶数)

Пример списка: `nums = [4, 7, 2, 9, 10, 13, 6]`

Ожидаемый вывод: `4`

Решение：

```python
nums = [4, 7, 2, 9, 10, 13, 6]

count = 0
for x in nums:
    if x % 2 == 0:
        count += 1

print(count)
```

#### Задание 4. Найти сумму всех положительных чисел в списке (求列表中所有正数的和)

Пример списка: `a = [-2, 3, 5, -7, 10]`

Ожидаемый результат: `18`

Решение：

```python
a = [-2, 3, 5, -7, 10]

total = 0
for x in a:
    if x > 0:
        total += x

print(total)
```

#### Задание 5. Найдите максимальную цифру в числе (找到一个整数中最大的数字)

Пример ввода: `583902`

Вывод: `9`

Решение：

```python
num = 583902

max_digit = 0
for ch in str(num):   # str() превращаем число в строку
    digit = int(ch)   # int() превращаем строку в число
    if digit > max_digit:
        max_digit = digit

print(max_digit)
```

#### Задание 6. Найтите самую длинную последовательность нулей (找到列表中连续 0 的最长长度)

Пример: `a = [1,0,0,2,0,0,0,5,0]` 

Вывод: '3'

```python
a = [1,0,0,2,0,0,0,5,0]

max_len = 0
current_len = 0

for x in a:
    if x == 0:
        current_len += 1
        max_len = max(max_len, current_len)
    else:
        current_len = 0

print(max_len)
```
