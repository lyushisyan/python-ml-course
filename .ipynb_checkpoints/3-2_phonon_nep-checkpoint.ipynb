### Построение фононного спектра на основе NEP (Neuroevolution Potential) 基于NEP势函数绘制声子谱

#### 1. Настройка среды и импорт библиотек (环境配置与库导入) 

1.1 Импортируются библиотеки, необходимые для расчета фононного спектра (ASE, Calorine, Phonopy и др.). 导入计算声子谱所需的库（ASE, Calorine, Phonopy 等）

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
from ase.io import read
from calorine.calculators import CPUNEP
from calorine.tools import get_force_constants, relax_structure
from seekpath import get_explicit_k_path
```

1.2 Настраивается стиль графиков (设置绘图风格)

```python
plt.rcParams.update({
    'font.family': 'sans-serif',           
    'font.sans-serif': ['Arial', 'Helvetica', 'DejaVu Sans'],
    'font.size': 14,                       
    'axes.linewidth': 1.5,                 
    'axes.labelsize': 16,                  
    'xtick.major.size': 6,                 
    'xtick.major.width': 1.5,              
    'xtick.direction': 'in',               
    'ytick.major.size': 6,
    'ytick.major.width': 1.5,
    'ytick.direction': 'in',
    'lines.linewidth': 1.5,                
})
```

#### 2. Конфигурация параметров (参数设置)
- Здесь настраиваются имена входных файлов, размер суперячейки (Supercell) и диапазон частот.
- 在此配置输入文件名、扩胞大小（Supercell）以及频率显示的范围。

```python
POSCAR_FILE = 'POSCAR-Si'            # 结构文件 / Файл структуры
NEP_MODEL_FILE = 'Si_NEP.txt'        # NEP 模型 / Модель NEP
SUPERCELL_MATRIX = [4, 4, 4]         # 扩胞大小 / Размер суперячейки
DOS_MESH = [50, 50, 50]              # DOS 网格 / Сетка DOS
FREQ_MAX = 16                        # 频率上限 / Макс. частота (THz)
TRUNCATE_LABEL = 'L'                 # 路径截断点 / Точка отсечения пути
```

#### 3. Оптимизация структуры и расчет фононов (结构优化与声子计算)
- Чтение структуры и загрузка модели машинного силового поля NEP.
- 读取结构并加载 NEP 机器学习力场模型。

```python
print(f"Reading structure from {POSCAR_FILE}...")
structure = read(POSCAR_FILE)
calculator = CPUNEP(NEP_MODEL_FILE)
structure.calc = calculator
```

- Релаксация структуры: Гарантирует, что система находится в минимуме энергии, иначе в спектре появятся «мнимые» частоты.
- 结构弛豫：确保体系处于能量最低点，否则声子谱会出现虚频。

```python
print("Relaxing structure...")
relax_structure(structure, fmax=0.0001)
```

- Расчет силовых констант: Получение матрицы силовых констант второго порядка методом смещений.
- 计算力常数：通过位移法获取二阶力常数矩阵。

```python
print(f"Computing force constants (Supercell: {SUPERCELL_MATRIX})...")
phonon = get_force_constants(structure, calculator, SUPERCELL_MATRIX)
```

- Генерация K-пути: Автоматическое определение пути через высокосимметричные точки с помощью seekpath.
- 生成路径：使用 seekpath 自动确定高对称 K 点路径。

```python
print("Generating k-path...")
structure_tuple = (structure.cell, structure.get_scaled_positions(), structure.numbers)
path_data = get_explicit_k_path(structure_tuple)

kpath_rel = path_data['explicit_kpoints_rel']           
kpath_linear = path_data['explicit_kpoints_linearcoord'] 
kpath_labels = path_data['explicit_kpoints_labels']
```

```python
if TRUNCATE_LABEL and (TRUNCATE_LABEL in kpath_labels):
    try:
        cut_index = kpath_labels.index(TRUNCATE_LABEL)
        kpath_rel = kpath_rel[:cut_index+1]
        kpath_linear = kpath_linear[:cut_index+1]
        kpath_labels = kpath_labels[:cut_index+1]
        print(f"Path truncated at '{TRUNCATE_LABEL}'.")
    except ValueError:
        pass
```

#### 4. Расчет фононного спектра и плотности состояний (DOS) (声子谱与态密度计算)
- Фононный спектр: Расчет частот фононов вдоль высокосимметричного пути.
- 声子谱：计算沿着高对称路径的声子频率。

```python
phonon.run_band_structure([kpath_rel], with_eigenvectors=False, labels=kpath_labels)
band_dict = phonon.get_band_structure_dict()
frequencies = band_dict['frequencies'][0]
```

- DOS: Расчет плотности фононных состояний по всей зоне Бриллюэна.
- DOS：计算全布里渊区的声子态密度。

```python
print("Computing DOS...")
phonon.run_mesh(DOS_MESH)
phonon.run_total_dos(freq_min=0, freq_max=FREQ_MAX*1.1, freq_pitch=0.05)
dos_dict = phonon.get_total_dos_dict()
dos_freqs = dos_dict['frequency_points']
total_dos = dos_dict['total_dos']
```

#### 5. Визуализация результатов (结果可视化)
- Наконец, мы строим график фононной дисперсии (слева) и плотности состояний (справа) на одном рисунке.
- 最后，我们将声子色散谱（左图）和态密度（右图）合并绘制在一张图中。

```python
print("Plotting with Publication Style...")

# Создание полотна: 2 подграфика с соотношением ширины 2:1 и общей осью Y
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5), 
                               gridspec_kw={'width_ratios': [2, 1]}, 
                               sharey=True)

# === Левый график: Дисперсия (Dispersion) ===
n_bands = frequencies.shape[1]
for i in range(n_bands):
    # Отрисовка фононных зон синим цветом
    ax1.plot(kpath_linear, frequencies[:, i], color='#1f77b4', alpha=0.9)

# Обработка высокосимметричных точек (вертикальные разделители)
tick_pos = []
tick_labs = []
for pos, lab in zip(kpath_linear, kpath_labels):
    if lab: 
        lab_str = r'$\Gamma$' if lab == 'GAMMA' else lab
        tick_pos.append(pos)
        tick_labs.append(lab_str)
        # Добавление черных вертикальных линий
        ax1.axvline(x=pos, color='k', linestyle='-', linewidth=1.0)

# Настройка оси X
ax1.set_xticks(tick_pos)
ax1.set_xticklabels(tick_labs)
ax1.set_xlim(kpath_linear[0], kpath_linear[-1])

# Настройка оси Y и подписей
ax1.set_ylabel('Frequency (THz)')
ax1.set_xlabel('Wave Vector')
ax1.set_ylim(0, FREQ_MAX)

# === Правый график: Плотность состояний (DOS) ===
# Отрисовка контура DOS черной линией
ax2.plot(total_dos, dos_freqs, 'k-')
# Заполнение области под кривой серым цветом
ax2.fill_betweenx(dos_freqs, 0, total_dos, color='gray', alpha=0.3)

# Настройка оси DOS
ax2.set_xlabel('DOS (a.u.)')
ax2.set_xlim(left=0)

# Скрытие меток оси Y для правого графика (используются метки левого)
plt.setp(ax2.get_yticklabels(), visible=False)

# Оптимизация расположения элементов и вывод на экран
plt.tight_layout()
plt.show()
```
